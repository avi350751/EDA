```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('world_population_data.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
sns.histplot(df['2023 population'],bins=10,kde=True)
plt.title('2023 Population distribution')
plt.show()
```

```python
df.groupby('continent')['2023 population'].mean().sort_values(ascending=False)
```

Highest population is from Asia and South America

```python
top_10_countries = df.sort_values(by='2023 population', ascending = False).head(10)
top_10_countries
```

```python
sns.barplot(x='2023 population',y='country',data=top_10_countries)
plt.title('Top 10 countries by population - 2023')
plt.xlabel('2023 population')
plt.ylabel('countries')
plt.show()
```

```python
df.columns
```

```python
df.rename(columns={'density (km²)':'density_km2'},inplace=True)
```

```python
df.head()
```

```python
sns.scatterplot(x='2023 population',y='density_km2',data=df)
plt.title('population density per sq km - 2023')
plt.xlabel('2023 population')
plt.ylabel('density')
plt.show()
```

```python
df.columns
```

```python
df.rename(columns={'area (km²)':'area_km2'},inplace=True)
```

```python
df.groupby('continent')['area_km2'].sum().sort_values(ascending=False)
```

```python
sns.violinplot(x='continent',y='density_km2',data=df,palette='viridis')
plt.title('Population density across continents')
plt.xlabel('Continent')
plt.ylabel('Population density')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.boxplot(x='continent',y='density_km2',data=df,palette='muted')
plt.title('Population density across continents')
plt.xlabel('Continent')
plt.ylabel('Population density')
plt.xticks(rotation=90)
plt.show()
```
