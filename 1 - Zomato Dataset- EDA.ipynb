```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
final_df.head()
```

```python
final_df.info()
```

### Observation

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_values = final_df.Country.value_counts().values
country_values
```

```python
# Set the figure size
plt.figure(figsize=(8, 10))
explode = [0.1 if i == 0 else 0 for i in range(15)]
plt.pie(country_values,labels=country_names,labeldistance=1.2,explode=explode)
plt.show()
```

```python
#For top 5 countries using Zomato

# Set the figure size
plt.figure(figsize=(5,5))
explode = [0.1 if i == 0 else 0 for i in range(5)]
plt.pie(country_values[:5],labels=country_names[:5],labeldistance=1.2,explode=explode,autopct='%1.2f%%')
plt.show()
```

Maximum number of transactions are from India, followed by US and UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

1. When rating is beyond 4.5 , experience is excellent
2. When ratings are between 4.0-4.4 , experience is very good
3. When ratings are between 3.5-3.9, experience is good
4. When ratings are between 2.5-3.4 , experience is average
5. When ratings are below 2.5 then experience is poor
6. Most number of customers have not rated the experience.

```python
matplotlib.rcParams['figure.figsize'] = (10,5)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
matplotlib.rcParams['figure.figsize'] = (15,12)
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','black'])
```

1. Not rated is high
2. Most of the ratings are between 2.5-3.9

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','black'])
```
